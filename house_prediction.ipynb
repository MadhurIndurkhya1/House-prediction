```python
data.drop(index= data[data['size'].isna()].index,inplace=True)
```

```python
data['bath'].fillna(data['bath'].mean(), inplace=True)
```

```python
data['bhk']= data['size'].str.split().str.get(0).astype(int)
```

```python
data['total_sqft'].unique()
```

```python
def convertRange(x):
  temp = x.split('-')
  if len(temp) == 2:
    return (float(temp[0]) + float(temp[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
data['total_sqft']= data['total_sqft'].apply(convertRange)
```

```python
data['location'].value_counts()
```

```python
data.describe()
```

```python
data.head()
```

```python
data.drop(columns=['size'],inplace=True)
```

```python
data['price_per_sqft'] = data['price']*100000 / data['total_sqft']
```

```python
data['location']= data['location'].fillna('').apply(lambda x: x.strip())
location_count = data['location'].value_counts()
```

```python
location_count_less_10 = location_count[location_count<=10]
location_count_less_10
```

```python
data['location']=data['location'].apply(lambda x: 'other' if x in location_count_less_10 else x)
```

```python
data.describe()
```

```python
data = data[((data['total_sqft'])/data['bhk'])>=300]
data.describe()
```

```python
def outlier_remover(df):
  df_output = pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)

    st = np.std(subdf.price_per_sqft)
    gen_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
    df_output = pd.concat([df_output,gen_df],ignore_index=True)
  return df_output
data = outlier_remover(data)
data.describe()
```

```python
def bhk_outlier_remover(df):
  exclude_indices = np.array([])
  for location, location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk] = {
          'mean': np.mean(bhk_df.price_per_sqft),
          'std': np.std(bhk_df.price_per_sqft),
          'count': bhk_df.shape[0]
      }
      for bhk, bhk_df in location_df.groupby('bhk'):
        stats = bhk_stats.get(bhk-1)
        if stats and stats['count']>5:
          exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<=(stats['mean'])].index.values)
  return df.drop(exclude_indices,axis='index')
```

```python
data = bhk_outlier_remover(data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso ,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x = data.drop(columns=['price'])
y = data['price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
colomn_trans = make_column_transformer((OneHotEncoder(sparse_output=False),['location']),
                                    remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(colomn_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
ls = Lasso()
```

```python
pipel = make_pipeline(colomn_trans,scaler,ls)
```

```python
pipel.fit(x_train,y_train)
```

```python
y_pred_ls = pipel.predict(x_test)
r2_score(y_test,y_pred_ls)
```

```python
rs = Ridge()
```

```python
piper = make_pipeline(colomn_trans,scaler,rs)
```

```python
piper.fit(x_train,y_train)
```
